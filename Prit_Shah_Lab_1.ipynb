#2. Import the q1data.csv dataset.

```python
import pandas as pd
import numpy as np
from plotnine import ggplot,aes,scale_size_continuous,geom_point,labs,geom_boxplot,scale_x_continuous,scale_y_continuous,scale_fill_manual,ggtitle,theme_void, geom_text, theme_light
```

```python
file_path = "/content/q1data.xlsm"

q1data = pd.read_excel(file_path)
print(q1data)
```

#3. Recreate the plot as best you can using plotnine.

```python
# Recreate the plot using plotnine
plot1 = (
    ggplot(q1data, aes(x="income", y="life_exp", fill="four_regions", size="population")) +
    geom_point() +
    theme_light()+
    scale_size_continuous(range=[1, 20], name="Population") +
    scale_x_continuous(
        name="Income",
        breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],
        labels=["500", "1000", "2000", "4000", "8000", "16k", "32k", "64k"],
        limits=[325, 128000],
        trans="log2"
    ) +
    scale_y_continuous(
        name="Life expectancy",
        breaks=list(range(20, 91, 10)),
        labels=["20", "30", "40", "50", "60", "70", "80", "90"],
        limits=[15, 90],
    ) +
    ggtitle("Life expectancy and income by region")+
    scale_fill_manual(
        values=["turquoise", "limegreen", "salmon", "yellow"],
        name="World regions",
        breaks=["africa", "americas", "asia", "europe"],
        labels=["Africa", "Americas", "Asia", "Europe"]
    ) +
    geom_text(aes(x=11000, y=50, label="2010"), size=100, alpha=0.002, color="lightgray")
)
plot1
```

#4. What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why its good or bad for visualizing these data.


```python
plot = (
    ggplot(q1data,aes(x = "income", y = "life_exp", fill = "four_regions")
           ) +
    geom_boxplot() +
    theme_light()+
    labs(
        title = "Life Expectancy and Income by World Region", x = "Income", y = "Life Expectancy", fill = "World Regions")
)
plot
```

# Substituting a different geometry for `geom_point` may typically lead to a loss of essential information, particularly regarding population size, which is represented by the relative sizes of the data points, in `geom_point`.

# "We could use `geom_boxplot` here alongside `geom_point`. The previous plot demonstrates the use of `geom_boxplot` as an additional geometry.

# With `box_plot`, we cannot observe the individual data points but we can provide a small outcome of each region. In the `box_plot` of the European region, the income and life expectancy are significantly higher when compared to all other regions.
